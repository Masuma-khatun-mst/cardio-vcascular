```python
hue_order = [0, 1]
hue_labels = ['No Cardiovascular Disease', 'Cardiovascular Disease']

palette = "dark" 
seaborn.countplot(data=df, x='gender', hue='cardio', palette=palette,hue_order=hue_order)

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Cardiovascular Disease Count by Gender')


plt.legend(title='Cardio',labels=hue_labels, loc='upper right')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rclf=RandomForestClassifier()
```

```python
rclf.fit(X_train,y_train)
```

```python
print("Performance: {:.2f}%".format(rclf.score(X_test,y_test)*100))
```
